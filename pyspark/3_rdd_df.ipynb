# RDD -> DF

# Summary
✔ RDD → DataFrame using .toDF()

✔ RDD → DataFrame using spark.createDataFrame()

✔ JSON → DataFrame using spark.read.json()

✔ Flattening nested RDDs before conversion

✔ Schema assignment using column names

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("RDDtoDF_SQL").getOrCreate()
sc = spark.sparkContext
```

# 1️⃣ Numbers → RDD → DataFrame → SQL (Sum, Avg)
## RDD

```python
numbers = [10, 20, 30, 40, 50]
rdd = sc.parallelize(numbers)
```

## DataFrame

```python
df = rdd.map(lambda x: (x,)).toDF(["value"])
df.show()
```

## Register SQL table

```python
df.createOrReplaceTempView("numbers")
```

## SQL query

```python
spark.sql("""
    SELECT 
        COUNT(*) AS count,
        SUM(value) AS total,
        AVG(value) AS average
    FROM numbers
""").show()
```

# 2️⃣ Tuples → RDD → DataFrame → SQL (Filter, Order)

## RDD

```python
students = [("Alice", 85), ("Bob", 72), ("Charlie", 90)]
rdd_students = sc.parallelize(students)
rdd_students.collect()
```

## Convert to DataFrame

```python
df_students = rdd_students.toDF(["name", "score"])
df_students.show()
```

# 3️⃣ List of Dicts → RDD → DataFrame

## RDD

```python
employees = [
    {"name": "John", "dept": "IT", "salary": 90000},
    {"name": "Sara", "dept": "HR", "salary": 75000},
    {"name": "Mike", "dept": "Finance", "salary": 80000}
]

rdd_employees = sc.parallelize(employees)
```

## Convert to DataFrame

```python
df_employees = spark.createDataFrame(rdd_employees)
df_employees.show()
```

# 4️⃣ JSON Strings → RDD → DataFrame

```python
json_data = [
    '{"city": "Boston", "temp": 75}',
    '{"city": "Chicago", "temp": 68}',
    '{"city": "Miami", "temp": 90}'
]

rdd_json = sc.parallelize(json_data)
```

## Convert to DataFrame

```python
df_json = spark.read.json(rdd_json)
df_json.show()
```

# 5️⃣ Nested List → RDD → Flatten → DataFrame

## RDD

```python
nested = [
    ["apple", "banana"],
    ["carrot", "beetroot"],
    ["mango", "orange"]
]

rdd_nested = sc.parallelize(nested)
```

## Flatten + Convert

```python
rdd_flat = rdd_nested.flatMap(lambda x: x)
df_flat = rdd_flat.map(lambda x: (x,)).toDF(["item"])
df_flat.show()
```
