## Python object → RDD (Spark)

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("RDDExamples").getOrCreate()
sc = spark.sparkContext
```

# RDD creation + transformation :-

# 1️⃣ Create RDD from a Python List (Basic Transformation)

```python
numbers = [1, 2, 3, 4, 5]
rdd_numbers = sc.parallelize(numbers)

rdd_squared = rdd_numbers.map(lambda x: x * x)

print(rdd_squared.collect())
```

# 2️⃣ Create RDD from a List of Tuples (Filter + Map)

```python
students = [("Alice", 85), ("Bob", 72), ("Charlie", 90), ("David", 60)]
rdd_students = sc.parallelize(students)

# Filter students scoring above 75
rdd_top = rdd_students.filter(lambda x: x[1] > 75)

# Add grade category
rdd_with_grade = rdd_top.map(lambda x: (x[0], x[1], "PASS"))

print(rdd_with_grade.collect())
```

# 3️⃣ Create RDD from a List of Dicts (Extract Fields)

```python
employees = [
    {"name": "John", "dept": "IT", "salary": 90000},
    {"name": "Sara", "dept": "HR", "salary": 75000},
    {"name": "Mike", "dept": "Finance", "salary": 80000}
]


rdd_employees = sc.parallelize(employees)

# Extract name and salary
rdd_salary = rdd_employees.map(lambda x: (x["name"], x["salary"]))

print(rdd_salary.collect())
```

# 4️⃣ Create RDD from JSON Strings (Parse + Transform)

```python
json_data = [
    '{"city": "Boston", "temp": 75}',
    '{"city": "Chicago", "temp": 68}',
    '{"city": "Miami", "temp": 90}'
]

import json

rdd_json = sc.parallelize(json_data)

# Parse JSON and convert temp to Celsius
rdd_parsed = rdd_json.map(lambda x: json.loads(x))

rdd_celsius = rdd_parsed.map(lambda x: (x["city"], (x["temp"] - 32) * 5/9))

print(rdd_celsius.collect())
```

# 5️⃣ Create RDD from a Nested Python Structure (Flattening)

```python
nested = [
    ["apple", "banana"],
    ["carrot", "beetroot"],
    ["mango", "orange"]
]

rdd_nested = sc.parallelize(nested)

# Flatten list of lists
rdd_flat = rdd_nested.flatMap(lambda x: x)

print(rdd_flat.collect())
```
