# 🌤️ Example 1 — Weather API (Open-Meteo)
Public, free, no API key required.

We will:
- Fetch hourly temperature data
- Parse JSON
- Convert to DataFrame
- Extract specific fields

**NOTE:** Run the below in a browser:

https://api.open-meteo.com/v1/forecast?latitude=40.71&longitude=-74.01&hourly=temperature_2m


````python
import requests
import pandas as pd

url = "https://api.open-meteo.com/v1/forecast"
params = {
    "latitude": 40.71,   # NYC
    "longitude": -74.01,
    "hourly": "temperature_2m,relativehumidity_2m"
}

response = requests.get(url, params=params)
data = response.json()

data
````

````python
df_weather = pd.DataFrame({
    "time": data["hourly"]["time"],
    "temperature": data["hourly"]["temperature_2m"],
    "humidity": data["hourly"]["relativehumidity_2m"]
})

df_weather.head()
````

### ✔ Use Case
- Build dashboards
- Weather alerts
- ETL pipelines for climate analytics


# 💰 Example 2 — Bitcoin Price API (CoinDesk)
We will:
- Fetch BTC price
- Parse nested JSON
- Extract USD price + timestamp
- Convert to DataFrame


````python
import requests
import pandas as pd

url = "https://api.coingecko.com/api/v3/simple/price"
params = {
    "ids": "bitcoin",
    "vs_currencies": "usd",
    "include_last_updated_at": "true"
}

response = requests.get(url, params=params)
btc_data = response.json()

btc_data
````

````python
btc_df = pd.DataFrame([{
    "time": pd.to_datetime(btc_data["bitcoin"]["last_updated_at"], unit="s"),
    "btc_usd": btc_data["bitcoin"]["usd"]
}])

btc_df
````

### ✔ Use Case
- Crypto dashboards
- Automated trading bots
- ETL pipelines for financial analytics


# 🐙 Example 3 — GitHub API (List Repositories)
We will:
- Call GitHub REST API
- Parse repo list
- Extract name, stars, language
- Convert to DataFrame

No API key needed for public repos.


````python
github_user = "torvalds"  # Example user
url = f"https://api.github.com/users/{github_user}/repos"

response = requests.get(url)
repos = response.json()

repos[:2]  # show first 2 repos
````

````python
df_repos = pd.DataFrame([
    {
        "name": repo.get("name"),
        "stars": repo.get("stargazers_count"),
        "language": repo.get("language")
    }
    for repo in repos
])

df_repos.head()
````

### ✔ Use Case
- Developer analytics
- Repo monitoring dashboards
- ETL pipelines for GitHub metadata
