## Business Requirement Summary — Interactive Country Filter for Aircraft Data
### Objective
```
Enable aviation analysts, operations teams, or data engineers to interactively explore live aircraft data by filtering flights based on their origin country. This supports real‑time situational awareness, regional monitoring, and operational decision‑making.
```

### Business Problem
```
Airspace monitoring teams often need to:

focus on a specific country’s aircraft

analyze regional congestion

identify unusual traffic patterns

support national aviation authorities

perform country‑specific operational reporting

Static charts are not enough — analysts need dynamic, user‑driven filtering inside the dashboard or notebook.
```

## What the dropdown displays when a country is selected
#### When you pick a country from the dropdown (for example: United States, Germany), the callback function produces three outputs, in this exact order:
```
1. A summary line
This tells you how many aircraft in the dataset originate from that country.

2. A preview table (first 5 rows)

3. A scatter plot (aircraft positions map)
A matplotlib chart showing longitude vs latitude for aircraft from that selected country.
Instead, it is a simple scatter plot:

✔ X‑axis = Longitude
✔ Y‑axis = Latitude
✔ Each dot = one aircraft
✔ The dots show where aircraft are located on Earth
✔ The plot updates when you select a country

Every aircraft has:
    a latitude (north/south position)
    a longitude (east/west position)
```

````python
from IPython.display import display, clear_output
import ipywidgets as widgets
import matplotlib.pyplot as plt

country_dropdown = widgets.Dropdown(
    options=sorted(df["origin_country"].dropna().unique()),
    description='Select Country:',
    value='United States',
    style={'description_width': 'initial'}
)

output_area = widgets.Output()

def update_country_view(change):
    with output_area:
        clear_output(wait=True)

        country = change['new']
        df_country = df[df["origin_country"] == country].copy()

        print(f"\nTotal aircraft from {country}: {len(df_country)}")
        display(df_country.head())

        if df_country.empty:
            print(f"No aircraft found for {country}.")
        else:
            plt.figure(figsize=(10,6))
            plt.scatter(
                df_country["longitude"],
                df_country["latitude"],
                s=12,
                alpha=0.6,
                color="green"
            )
            plt.title(f"Aircraft Positions — {country}")
            plt.xlabel("Longitude")
            plt.ylabel("Latitude")
            plt.grid(True)
            plt.show()

country_dropdown.observe(update_country_view, names='value')

display(country_dropdown, output_area)
````
