### Swapping - using a TEMP variable

````python
# 1. SWAPPING
print("Before Swapping::--")
# Assignment is done only once
left = 10 # ESPN
right = 5 # Fox News
print(f"left={left}, right: {right}") # left:10  right:5

# SWAPPING -- Write the LOGIC
temp = left # 10
left = right # 5
right = temp # 10

# After swapping, we need to see the below input. SO, how -- write the logic
print("\n After Swapping:--")
print(f"left={left}, right: {right}")  # left:5 right:10 (EXPECTED)
````

````python
# 1. SWAPPING
print("Before Swapping::--")
# Assignment is done only once
left = "ESPN"
right = "Fox News"
print(f"left={left}, right: {right}") # left:10  right:5

# SWAPPING -- Write the LOGIC (Transformation)
temp = left # 10
left = right # 5
right = temp # 10

# After swapping, we need to see the below input. SO, how -- write the logic
print("\n After Swapping:--")
print(f"left={left}, right: {right}")  # left:5 right:10 (EXPECTED)
````

### Space Complexity, Time Complexity

# OPTION_1: How many Variables: THREE (i.e. using a TEMP)
```
Storage (variables): 3
Compute : Operators = 3 *** Good - takes less compute/operations/less time ***
```

````python
# OPTION_1: How many Variables: THREE (i.e. TEMP)
# Each number takes 4 bytes per variable. Since we have 3 variable: 3 * 4 bytes = 12 bytes(memory the prog is taking)
a, b = 10, 5
temp = a; a = b; b = temp
print(a, b)
````

# OPTION_2: WITHOUT a temp variable - HOW?
```
- OPERATORS: we used in total 6 operations.
Storage (variables): 2 *** Good - takes less memory/storage ***
Operators = 6
```

````python
a, b = 10, 5 # current state of a, b : 10, 5

a = a + b # a=15; b=5 (New state at that point in time)
b = a - b # a=15; b=10 (New state, here, we populated the right value in 'b')
a = a - b

print(a, b)
````

````python
# Option-3: Operations: 2; Storage:2
# How to do swapping for STRINGS - without using TEMP variable
a = "ESPN"
b = "Fox News"
print(a, b)

a, b = b, a

print(a, b)
````

## Sorting -> makes use -> swapping

````python
in_lst = [ 5 , 3, 2, 4, 1] # Not sorted; not arranged in increasing order


# Sorting
#1. iterate thru each num in the list
for i in range(len(in_lst)):
    print(f"--> {i}, {in_lst[i]}")

# Expected output: [ 1, 2, 3, 4, 5]
````

````python
# How swapping is used within the sorting.
# How to use Nested-Loops ie. 'For' within a 'For'
````

````python
in_lst = [ 5 , 3, 2, 4, 1] # Not sorted; not arranged in increasing order

# Sorting
#1. iterate thru each num in  the list
for j in range(len(in_lst) - 1):       # j : 0,1,2,3,4
    
    for i in range(0, len(in_lst) - j-1):#
        # if first number is larger(5) then SWAP
        if in_lst[i] > in_lst[i+1]:
            temp = in_lst[i] ; in_lst[i]  = in_lst[i+1]; in_lst[i+1] = temp
            # print(f"--> {i}, {in_lst}")
            
    # print(f"===> in_lst: {in_lst}")

print(in_lst)
# Expected output: [ 1, 2, 3, 4, 5]
````
