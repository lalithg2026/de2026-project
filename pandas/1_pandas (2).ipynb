# Reading a CSV File in Python
We create a sample CSV and read it using Python and pandas.

```python
import pandas as pd

# --- Sample CSV Data --- Hardcoding/Stubbing/Manually creating the Data.
csv_data = """
id,name,age,city
1,Alice,25,New York
2,Bob,30,Los Angeles
3,Charlie,35,Chicago
4,David,40,Houston
5,Alice1,35,New York
"""

# Write CSV to a file
with open("sample_data.csv", "w") as f:
    f.write(csv_data)

# Read CSV -> loaded the data into a data frame
df = pd.read_csv("sample_data.csv")
# type(df)
df
```

# 📘 15 Essential Pandas Operations

```python
# 1️⃣ head() — View first rows
df.head(2)

# df.tail()
```

```python
# 2️⃣ info() — Summary of columns & types
df.info()
```

```python
# 3️⃣ describe() — Statistical summary
df.describe()
```

```python
# 4️⃣ df['col'] — Select a column
# df
df['age']
```

```python
# 5️⃣ df[['col1','col2']] — Select multiple columns
df[['name', 'city']]
```

```python
# 6️⃣ Filtering rows — df[df['age'] > 30]
# df['age'] # ALL the values present in that column
# df['age'] > 30 # Returns if the condition is True/False. All rows returned.
df[df['age'] > 30] # Only rows meeting the criteria are fetched. FILTERED rows. *****
```

```python
# 7️⃣ sort_values() — Sort rows - REARRANGE the Rows. The is no loss of data.
# df
# df.sort_values(by='age', ascending=True)
df.sort_values(by='city', ascending=False)
```

```python
# 8️⃣ groupby() — Group and aggregate
df
df.groupby('city')['age'].mean()
```

```python
# 9️⃣ assign() — Add new calculated column
# How to assign a default value to a NEW column:-
# df.assign(age_plus_10 = '')
# df.assign(age_plus_10 = 0)
# df.assign(age_plus_10 = True)

################################################
# TRANSFORMATION - calculated/processed/Modified
# How to ADD a NEW column
################################################
# How to assign based on an EXISTING column value -> assign to a NEW column:-
df1 = df.assign(age_plus_10 = df['age'] + 10)
df
df1
```

```python
# Example for rename from documentation - way to learn
df1 = pd.DataFrame({"A": [1, 2, 3], "B": [4, 5, 6]})
df1
df1.rename(columns={"A": "a", "B": "c"})
df1
```

```python
# 🔟 rename() — Rename columns
df.rename(columns={'name': 'full_name'})
```

```python
# 1️⃣1️⃣ drop() — Remove columns
df.drop(columns=['city'])
```

```python
# 1️⃣2️⃣ fillna() — Replace missing values
# df.fillna({'city': 'Unknown'})
df.fillna({'age': 0, 'city': 'Unknown'})
```

```python
# 1️⃣3️⃣ isnull() — Check missing values
df.isnull().sum()
```

```python
# 1️⃣4️⃣ value_counts() — Frequency count
df
df['city'].value_counts()
```

```python
# 1️⃣5️⃣ to_csv() — Save DataFrame to CSV
df.to_csv('output_0630.csv', index=False)
df
```
