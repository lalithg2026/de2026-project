# 🛒 Retail Sales ETL & Data Analysis (Beginner-Friendly)
This notebook demonstrates a **real-world ETL pipeline** using pandas.

Requirement:
- Create sample CSV files (Sales + Product Lookup)
- Read CSV files (Extract)
- Clean, transform, enrich data (Transform)
- Save final dataset (Load)
- Perform useful data analysis

This is a realistic scenario used in **data engineering** and **analytics**.

# 📘 1. Create Sample CSV Files
We generate two CSVs:

### **sales.csv**
Contains daily retail transactions.

### **products.csv**
Contains product master data (lookup table).

```python
import pandas as pd

# --- Sales CSV ---
sales_csv = """
sale_id,product_id,quantity,price,date
1,101,2,15.50,2024-01-01
2,103,1,45.00,2024-01-01
3,102,3,12.00,2024-01-02
4,101,1,15.50,2024-01-02
5,104,5,8.00,2024-01-03
"""

with open("sales.csv", "w") as f:
    f.write(sales_csv)

# --- Product Lookup CSV ---
products_csv = """
product_id,product_name,category
101,Notebook,Stationery
102,Pen,Stationery
103,Backpack,Bags
104,Water Bottle,Accessories
"""

with open("products.csv", "w") as f:
    f.write(products_csv)
```

# 📘 2. Extract — Read CSV Files
We load both CSVs into pandas DataFrames.

```python
sales_df = pd.read_csv("sales.csv")
products_df = pd.read_csv("products.csv")

sales_df, products_df
```

# 📘 3. Transform — Clean & Enrich Data
### Steps:
- Add **total_amount** column
- Join sales with product lookup
- Convert date column to datetime
- Sort by date
- Filter invalid rows (if any)

```python
# Add total_amount
sales_df['total_amount'] = sales_df['quantity'] * sales_df['price']

# Join with product lookup
merged_df = sales_df.merge(products_df, on='product_id', how='left')

# Convert date to datetime
merged_df['date'] = pd.to_datetime(merged_df['date'])

# Sort by date
merged_df = merged_df.sort_values(by='date')

merged_df
```

# 📘 4. Load — Save Final ETL Output
We save the enriched dataset to a new CSV.

```python
merged_df.to_csv("sales_etl_output.csv", index=False)
merged_df
```

# 📊 5. Data Analysis — Real Insights
We now perform useful analytics:

### 1. Total revenue
### 2. Revenue by category
### 3. Top-selling products
### 4. Daily sales summary

```python
# Total revenue
total_revenue = merged_df['total_amount'].sum()

# Revenue by category
revenue_by_category = merged_df.groupby('category')['total_amount'].sum()

# Top-selling products
top_products = merged_df.groupby('product_name')['quantity'].sum().sort_values(ascending=False)

# Daily sales summary
daily_summary = merged_df.groupby('date')['total_amount'].sum()

total_revenue, revenue_by_category, top_products, daily_summary
```
